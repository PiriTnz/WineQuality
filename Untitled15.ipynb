```python
#Essential libraries for data representation
import pandas as pd
import numpy as np

#Display and plotting libraries
from IPython.display import display
import seaborn as sn
import matplotlib.pyplot as plt

#Statistics libraries
from scipy import stats
from collections import Counter

#ML Libraries
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import OneHotEncoder

# For dimension reduction
from sklearn.manifold import TSNE
from sklearn.manifold import Isomap
from sklearn.decomposition import PCA

#DNN libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
```

```python
# prompt: ## Task 1: Importing and Feature Engineering
# 1. Create the corresponding Pandas dataframes `white_wine_df` and `red_wine_df`from `winequality-white.csv` and `winequality-red.csv` respectively. Print the shape and the first `5` lines for each dataframe.
# 2. To simplify our task, we will aggregate the wine `quality` scores into low-quality (scores lower or equal to 5), medium-quality (scores equal to 6 or 7) and high-quality (scores higher or equal to 8) wines. Create a new column `quality_label` in the two previous dataframes with the corresponding values `low`, `medium` and `high`.
# 3. Add a new column `wine_type` in each dataframe containing the corresponding values `red` and `white`. Concatenate the red and white wines dataframes into a single `wines_df` dataframe. Shuffle the rows of the `wines_df` dataframe and reset its index.

# Create the corresponding Pandas dataframes white_wine_df and red_wine_df
white_wine_df = pd.read_csv('winequality-white.csv', sep=';')
red_wine_df = pd.read_csv('winequality-red.csv', sep=';')

# Print the shape and the first 5 lines for each dataframe.
print("Shape of white wine dataframe:", white_wine_df.shape)
display(white_wine_df.head())
print("Shape of red wine dataframe:", red_wine_df.shape)
display(red_wine_df.head())

# Aggregate the wine quality scores
white_wine_df['quality_label'] = white_wine_df['quality'].apply(lambda x: 'low' if x <= 5 else ('medium' if x <= 7 else 'high'))
red_wine_df['quality_label'] = red_wine_df['quality'].apply(lambda x: 'low' if x <= 5 else ('medium' if x <= 7 else 'high'))

# Add a new column wine_type
white_wine_df['wine_type'] = 'white'
red_wine_df['wine_type'] = 'red'

# Concatenate the dataframes
wines_df = pd.concat([red_wine_df, white_wine_df])

# Shuffle the rows and reset index
wines_df = wines_df.sample(frac=1).reset_index(drop=True)
```

```python
# prompt: ## Task 2: Exploratory Data Analysis
# 1. For each type of wine (red or white), print some descriptive statistics (using `describe`) for the columns `residual sugar, total sulfur dioxide, sulphates, alcohol, volatile acidity` and `quality`. What can you hypothesize?
# 2. For each quality of wine (`low, medium` and `high`), print some descriptive statistics for the columns `alcohol`, `volatile acidity`, `pH` and `quality`. What can you hypothesize?
# 3. We will use the one-way ANOVA test to compare the difference in average among subsets of data. Use the function `stats.f_oneway` from `scipy` to compare the `pH` and `alcohol` means for our three categories of wines (`low`, `medium` and `high`). If the p-value return is less than 0.05, we can infer that there is a statistically significant difference in mean levels. Do the results align with the previous question?
# 4. Visualise the alcohol and pH levels for the three wine quality classes (`low`, `medium` and `high`) using two boxplots.

# 1. Descriptive statistics for each wine type
for wine_type in ['red', 'white']:
    print(f"Descriptive statistics for {wine_type} wine:")
    subset = wines_df[wines_df['wine_type'] == wine_type]
    desc_stats = subset[['residual sugar', 'total sulfur dioxide', 'sulphates', 'alcohol', 'volatile acidity', 'quality']].describe()
    display(desc_stats)
    print("\n")

# Hypothesis:  We can hypothesize relationships between wine quality and characteristics like alcohol content, volatile acidity and sulphates. Higher alcohol might correlate with higher quality. Lower volatile acidity could also be associated with higher quality.

# 2. Descriptive statistics for each wine quality
for quality_label in ['low', 'medium', 'high']:
    print(f"Descriptive statistics for {quality_label} quality wine:")
    subset = wines_df[wines_df['quality_label'] == quality_label]
    desc_stats = subset[['alcohol', 'volatile acidity', 'pH', 'quality']].describe()
    display(desc_stats)
    print("\n")

# Hypothesis: Wines with higher quality tend to have higher alcohol content, lower volatile acidity and a certain pH range

# 3. One-way ANOVA test
for column in ['pH', 'alcohol']:
    low = wines_df[wines_df['quality_label'] == 'low'][column]
    medium = wines_df[wines_df['quality_label'] == 'medium'][column]
    high = wines_df[wines_df['quality_label'] == 'high'][column]
    f_statistic, p_value = stats.f_oneway(low, medium, high)
    print(f"One-way ANOVA test for {column}:")
    print(f"F-statistic: {f_statistic:.2f}, p-value: {p_value:.3f}")
    if p_value < 0.05:
        print(f"There is a statistically significant difference in the mean {column} across the quality labels")
    else:
        print(f"There is no statistically significant difference in the mean {column} across the quality labels")
    print("\n")

# 4. Boxplots
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sn.boxplot(x='quality_label', y='alcohol', data=wines_df)
plt.title('Alcohol Levels by Wine Quality')

plt.subplot(1, 2, 2)
sn.boxplot(x='quality_label', y='pH', data=wines_df)
plt.title('pH Levels by Wine Quality')

plt.tight_layout()
plt.show()
```

```python
# prompt: ## Task 3: Dimension Reduction
# 1. Use Scikit-learn functions `TSNE`, `Isomap`, and `PCA` to reduce the features to 2D points.
# 2. Create several plots for each method and colouration (different colours for wine types or different colours for wine quality). What can you say?

# Feature scaling
X = wines_df.drop(['quality', 'quality_label', 'wine_type'], axis=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dimensionality reduction
methods = {
    'TSNE': TSNE(n_components=2, random_state=42),
    'Isomap': Isomap(n_components=2),
    'PCA': PCA(n_components=2)
}

for method_name, method in methods.items():
    X_reduced = method.fit_transform(X_scaled)

    # Create plots colored by wine type
    plt.figure(figsize=(8, 6))
    for wine_type in wines_df['wine_type'].unique():
        plt.scatter(X_reduced[wines_df['wine_type'] == wine_type, 0],
                    X_reduced[wines_df['wine_type'] == wine_type, 1],
                    label=wine_type)
    plt.title(f'{method_name} - Colored by Wine Type')
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.legend()
    plt.show()

    # Create plots colored by wine quality
    plt.figure(figsize=(8, 6))
    for quality_label in wines_df['quality_label'].unique():
        plt.scatter(X_reduced[wines_df['quality_label'] == quality_label, 0],
                    X_reduced[wines_df['quality_label'] == quality_label, 1],
                    label=quality_label)
    plt.title(f'{method_name} - Colored by Wine Quality')
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.legend()
    plt.show()

# Analysis
# Based on the plots, observe the separation of wine types and quality levels in the reduced dimensions.
# TSNE often reveals clusters, while PCA gives a linear representation. Isomap preserves distances.
# Comment on the effectiveness of each method in visually separating the different categories.
```

```python
# prompt: ## Task 4: Making Predictions
# ### Predicting the wine type
# 1. Normalize the 11 physicochemical features of the `wines_df` dataset using the `StandardScaler`. Then, use `train_test_split` to split the dataset into a training and testing dataset containing respectively 70% and 30% of the original dataset. Print the number of red/white wines in each dataset. Fit the default Scikit-learn's `LogisticRegression` model on the training set and predict the wine types of the test set.
# 2. It is time to evaluate our regression model. Use `classification_report`, `confusion_matrix` from Scikit-learn's `metrics` module to show the result of the prediction. Print model performance metrics such as the accuracy, precision, recall and F1-score.
# ###

# Split data into training and testing sets
X = wines_df.drop(['quality', 'quality_label', 'wine_type'], axis=1)
y = wines_df['wine_type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Count red and white wines in each set
train_counts = Counter(y_train)
test_counts = Counter(y_test)

print("Training Set Counts:")
print(f"Red Wines: {train_counts['red']}")
print(f"White Wines: {train_counts['white']}")

print("\nTesting Set Counts:")
print(f"Red Wines: {test_counts['red']}")
print(f"White Wines: {test_counts['white']}")

# Train a Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

# Calculate and print additional metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label='white') #Example positive label
recall = recall_score(y_test, y_pred, pos_label='white') #Example positive label
f1 = f1_score(y_test, y_pred, pos_label='white') #Example positive label

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
```

```python
# prompt: ### Predicting the wine quality
# 1. Normalize the 11 physicochemical features of the `wines_df` dataset using the `StandardScaler`. Then, use `train_test_split` to split the dataset into a training and testing dataset containing respectively 70% and 30% of the original dataset. Print the number of low, medium and high-quality wines for each dataset. Fit the default Scikit-learn's `DecisionTreeClassifier` model on the training set and predict the wine quality of the test set.
# 2. Evaluate your wine quality model.
# 3. Use the default `RandomForestClassifier` from the Scikit-learn's `ensemble` module. Fit it on the training dataset and predict it on the test dataset. Evaluate the approach and compare it to the decision tree model that you created.

# 1. Predicting wine quality

# Normalize features and split data
X = wines_df.drop(['quality', 'quality_label', 'wine_type'], axis=1)
y = wines_df['quality_label']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Print quality counts
train_counts = Counter(y_train)
test_counts = Counter(y_test)
print("Training Set Counts:")
print(train_counts)
print("\nTesting Set Counts:")
print(test_counts)

# Decision Tree Classifier
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)

# 2. Evaluate Decision Tree model
print("\nDecision Tree Classifier Evaluation:")
print(classification_report(y_test, dt_predictions))
print(confusion_matrix(y_test, dt_predictions))

# 3. Random Forest Classifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)

# Evaluate Random Forest model and compare
print("\nRandom Forest Classifier Evaluation:")
print(classification_report(y_test, rf_predictions))
print(confusion_matrix(y_test, rf_predictions))

# Compare the two models (You can add more detailed comparison metrics)
print("\nModel Comparison:")
print(f"Decision Tree Accuracy: {accuracy_score(y_test, dt_predictions)}")
print(f"Random Forest Accuracy: {accuracy_score(y_test, rf_predictions)}")
```

```python
# prompt: 4. Print the parameters of your `RandomForestClassifier` using the `get_params` method. Choose a subset of parameters to vary (including `n_estimators` and `max_features`) and use Scikit-learn's `GridSearchCV` to select the best parameters with `cv=5`.
# 5. Compare the `GridSearchCV`'s best-obtained model with the default one.
# **BONUS: Create a three-hidden layer DNN using to predict the tasks above.

# 4. Print parameters and perform GridSearchCV
print(rf_model.get_params())

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_features': ['sqrt', 'log2'],
    # Add other parameters to tune
}

grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

print("Best parameters found by GridSearchCV:", grid_search.best_params_)
best_rf_model = grid_search.best_estimator_
best_rf_predictions = best_rf_model.predict(X_test)

# 5. Compare models
print("\nComparison of Default and Best Random Forest Models:")
print("Default Random Forest:")
print(classification_report(y_test, rf_predictions))
print("\nBest Random Forest (GridSearchCV):")
print(classification_report(y_test, best_rf_predictions))


# BONUS: Three-hidden layer DNN
model = Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax') # 3 output neurons for 'low', 'medium', 'high'
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',  # Use sparse_categorical_crossentropy for integer labels
              metrics=['accuracy'])

# Convert quality labels to numerical representation for DNN
label_mapping = {'low': 0, 'medium': 1, 'high': 2}
y_train_num = y_train.map(label_mapping)
y_test_num = y_test.map(label_mapping)


model.fit(X_train, y_train_num, epochs=50, batch_size=32, validation_split=0.2) # Adjust epochs and batch size
loss, accuracy = model.evaluate(X_test, y_test_num)
print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')

y_pred_dnn = np.argmax(model.predict(X_test), axis=1) # Get predicted class labels

# Convert numerical predictions back to original labels
reverse_mapping = {v: k for k, v in label_mapping.items()}
y_pred_dnn_labels = [reverse_mapping[pred] for pred in y_pred_dnn]


print("\nDNN Classifier Evaluation:")
print(classification_report(y_test, y_pred_dnn_labels))
print(confusion_matrix(y_test, y_pred_dnn_labels))
```
